```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
app_df = pd.read_csv("application_record.csv")
```

```python
app_df.head()
```

```python
app_df.info()
```

```python
app_df.isna().sum()
```

```python
app_df_null = app_df.dropna(axis=0)
app_df_null.isnull().sum()
```

```python
app_df_drop = app_df_null.drop(['DAYS_BIRTH','DAYS_EMPLOYED','FLAG_MOBIL','FLAG_PHONE','FLAG_WORK_PHONE','FLAG_EMAIL','OCCUPATION_TYPE'],axis=1)
app_df_drop.isnull().sum()
```

```python
app_df_drop.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
```

```python
cat_data = ['CODE_GENDER','FLAG_OWN_CAR','FLAG_OWN_REALTY','NAME_INCOME_TYPE','NAME_EDUCATION_TYPE','NAME_FAMILY_STATUS','NAME_HOUSING_TYPE']
```

```python
app_df_drop[cat_data] = app_df_drop[cat_data].apply(lb.fit_transform)
```

```python
app_df_drop
```

```python
#Checking Outliers & remove Outliers
high_bound = app_df_drop['CNT_CHILDREN'].quantile(0.999)
print('high_bound :', high_bound)
low_bound = app_df_drop['CNT_CHILDREN'].quantile(0.001)
print('low_bound :', low_bound)
```

```python
app_df_drop = app_df_drop[(app_df_drop['CNT_CHILDREN']>=low_bound) & (app_df_drop['CNT_CHILDREN']<=high_bound)]
```

```python
high_bound = app_df_drop['AMT_INCOME_TOTAL'].quantile(0.999)
print('high_bound :', high_bound)
low_bound = app_df_drop['AMT_INCOME_TOTAL'].quantile(0.001)
print('low_bound :', low_bound)
```

```python
app_df_drop = app_df_drop[(app_df_drop['AMT_INCOME_TOTAL']>=low_bound) & (app_df_drop['AMT_INCOME_TOTAL']<=high_bound)]
```

```python
high_bound = app_df_drop['CNT_FAM_MEMBERS'].quantile(0.999)
print('high_bound :', high_bound)
low_bound = app_df_drop['CNT_FAM_MEMBERS'].quantile(0.001)
print('low_bound :', low_bound)
```

```python
app_df_drop = app_df_drop[(app_df_drop['CNT_FAM_MEMBERS']>=low_bound) & (app_df_drop['CNT_FAM_MEMBERS']<=high_bound)]
```

```python
credit_df = pd.read_csv("credit_record.csv")
```

```python
credit_df.head()
```

```python
credit_df.isnull().sum()
```

```python
# categorizing 'STATUS' column to binary classification   0 : Good Client and 1 : bad client
credit_df['STATUS'].replace(['C', 'X'],0, inplace=True)
```

```python
credit_df['STATUS'].replace(['2','3','4','5'],1, inplace=True)
```

```python
credit_df['STATUS'] = credit_df['STATUS'].astype('int')
```

```python
credit_df.info()
```

```python
credit_df_trans = credit_df.groupby('ID').agg(max).reset_index()
credit_df_trans.drop('MONTHS_BALANCE', axis=1, inplace=True)
credit_df_trans.head()
```

```python
#adding to data
final_df = pd.merge(app_df_drop, credit_df_trans, on='ID', how='inner')
final_df.head()
```

```python
final_df.shape
```

```python
# dropping 'ID' column as it is having only unique values (not required for ML Model)
final_df.drop('ID', axis=1, inplace=True)
```

```python
# checking if there are still duplicate rows in Final Dataframe
len(final_df) - len(final_df.drop_duplicates())
```

```python
# Dropping duplicate records
final_df = final_df.drop_duplicates()
final_df.reset_index(drop=True ,inplace=True)
```

```python
final_df.shape
```

```python
final_df.isnull().sum()
```

```python
final_df.head()
```

```python
#Dependent & Independant Variable
X = final_df.iloc[:,0:-1]
X
```

```python
y = final_df.iloc[:,-1]
y
```

```python
#Feature Scaling
# from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
# sc.fit_transform(X)
from sklearn.preprocessing import MinMaxScaler
nm = MinMaxScaler()
nm.fit_transform(X)
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(nm.fit_transform(X),y,test_size=0.2,random_state = 10)
```

```python
#Using Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

lr = LogisticRegression()
lr.fit(X_train, y_train)

print('Logistic Model Accuracy : ', lr.score(X_test, y_test)*100, '%')

prediction = lr.predict(X_test)
print('\nConfusion matrix :')
print(confusion_matrix(y_test, prediction))
      
print('\nClassification report:')      
print(classification_report(y_test, prediction))
```

```python
# using SVM
from sklearn.svm import SVC

svc_model = SVC()

svc_model.fit(X_train, y_train)

print('Support Vector Classifier Accuracy : ', svc_model.score(X_test, y_test)*100, '%')

prediction = svc_model.predict(X_test)
print('\nConfusion matrix :')
print(confusion_matrix(y_test, prediction))
      
print('\nClassification report:')      
print(classification_report(y_test, prediction))
```

```python
print("Prediction table:")
a=svc_model.predict(X_test)
b=y_test
z=list(enumerate(a))
z=[x[0] for x in z]
table={'Index':z,'Predicted value':a,'Real value':b}
nf=pd.DataFrame(table)
nf.style.hide_index()
nf.head(75)

# nf['Predicted value'].unique()
```
